# **Dataset**

**credits** bao gồm các đặc trưng sau:

* **movie_id**: id của phim
* **cast**: tên của các diễn viên chính và phụ.
* **crew**: tên của các nhân viên trong quá trình sản xuất phim (đạo diễn, editor, người soạn nhạc, v.v.)

**movies** bao gồm các đặc trưng sau:

* **budget**: kinh phí thực hiện bộ phim.
* **genre**: các thể loại của phim.
* **homepage**: liên kết đến trang chủ của bộ phim.
* **id**: thông tin về movie_id như trong tập dữ liệu tmdb_5000_credits.
* **keywords**: các từ khóa hoặc thẻ liên quan đến bộ phim.
* **original_language**: ngôn ngữ trong bộ phim sử dụng.
* **original_title**: tên phim trước khi dịch hoặc chuyển thể.
* **overview**: mô tả ngắn gọn về bộ phim.
* **popularity**: số lượng chỉ định mức độ phổ biến của bộ phim.
* **production_companies**: nhà sản xuất của bộ phim.
* **production_countries**: quốc gia sản xuất của bộ phim.
* **release_date**: ngày mà nó được phát hành.
* **revenue**: doanh thu trên toàn thế giới do bộ phim tạo ra.
* **runtime**: thời gian  của phim tính bằng phút.
* **status**: Đã phát hành hoặc còn là tin đồn.
* **tagline**: khẩu hiệu của phim.
* **title**: tên phim.
* **vote_average**: đánh giá trung bình mà bộ phim nhận được.
* **vote_count**: số lượng đánh giá.

```python
# dataset
df_movies = pd.read_csv('./tmdb_5000_movies.csv')
df_credits = pd.read_csv('./tmdb_5000_credits.csv') 

# get columns
columns_rename = df_movies.columns.values
# rename 'id' column to 'movie_id' in df_movies
columns_rename[columns_rename == 'id'] = 'movie_id'
# assign new columns for df_movies
df_movies.columns = columns_rename

# merge dataframe by movie_id
df = df_movies.merge(df_credits, on='movie_id')
print("DF Merge shape: ", df.shape)
df.head()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

# **Content Based Filtering**

Trong hệ thống đề xuất này, nội dung của phim (overview, cast, crew, keyword, v.v.) được sử dụng để tìm sự tương đồng của giữa các bộ phim. Sau đó, những bộ phim có nhiều khả năng tương tự nhất được đề xuất.

Ta sẽ tính toán điểm tương đồng theo từng cặp cho tất cả các phim dựa trên mô tả cốt truyện của chúng và đề xuất các phim dựa trên điểm tương đồng đó.

Thay thế NaN bằng chuỗi trống: "".

```python
df['tags'] = df['tags'].fillna("")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import pickle
```

```python
# Tạo ma trận đặc trưng
cv = CountVectorizer(max_features=5000, stop_words='english')
vector = cv.fit_transform(df['tags']).toarray()
```

```python
vector.shape
```

### **Train/Test split**

```python
# Chia dữ liệu thành tập huấn luyện và tập kiểm thử
X_train, X_test = train_test_split(vector, test_size=0.2, random_state=42)
```

```python
# Tính toán ma trận tương tự cho cả tập huấn luyện và tập kiểm thử
similarity_train = cosine_similarity(X_train)
similarity_test = cosine_similarity(X_test)
```

### **Đánh giá mô hình dựa vào tính trung bình khoảng cách cosine trên tập huấn luyện và tập test**

```python
loss_train = 1 - similarity_train
average_cosine_distance = np.mean(loss_train)
print(f"Average Cosine Distance: {average_cosine_distance}")
```

```python
loss_test = 1 - similarity_test
average_cosine_distance = np.mean(loss_test)
print(f"Average Cosine Distance: {average_cosine_distance}")
```

```python
def recommend(movie):
    index = df[df['title_x'] == movie].index[0]
    distances = sorted(list(enumerate(similarity_test[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0]].title_x)
        
```

```python
# Kiểm tra kết quả trả về
recommend('Avatar')
```

### **Kiểm tra độ chính xác đối với phim Avatar**

```python
# Tìm chỉ số của phim 'Avatar' trong dataframe
avatar_index = df[df['title_x'] == 'Avatar'].index[0]

# Tính độ tương tự cosine similarity giữa 'Avatar' và các bộ phim so sánh
compare_movies = ['The Green Hornet', 'The Last Castle', 'Kangaroo Jack', 'The Shaggy Dog', 'Big Fish']
compare_indices = [df[df['title_x'] == movie].index[0] for movie in compare_movies]

similarities = [cosine_similarity([similarity_test[avatar_index]], [similarity_test[idx]])[0][0] for idx in compare_indices]

# In ra các kết quả
for movie, similarity in zip(compare_movies, similarities):
    print(f"Similarity between 'Avatar' and '{movie}': {similarity}")
```

## Single Value Decomposition

***`Nhược điểm của Lọc Cộng tác về khả năng mở rộng (scalability) và độ thưa thớt (sparsity)`***. Để xử lý vấn đề này chúng ta có thể sử dụng latent factor model (mô hình yếu tố tiềm ẩn) để nắm bắt sử giống nhau của người dùng và bộ phim.

Về cơ bản, chúng ta muốn biến vấn đề đề xuất thành vấn đề tối ưu hóa. ***`Chúng ta có thể xem đó là mức độ tốt của chúng ta trong việc dự đoán xếp hạng cho các phim do người dùng cung cấp. Với chi phí càng nhỏ thì dự đoán càng tốt.`***

Một latent factor model mà chúng ta có thể sử dụng là [Single Value Decomposition](https://machinelearningcoban.com/2017/06/07/svd/) mà chúng ta đã tìm hiểu từ [Matrix Factorization](https://machinelearningcoban.com/2017/05/31/matrixfactorization/)

## Single Value Decomposition

Ở phần này chúng ta sẽ sử dụng phương pháp lọc công tác Matrix Factorization bằng phương pháp tối ưu Single Value Decomposition với chi phí càng nhỏ thì càng tối ưu.

Tham khảo đường dẫn sau: [Use custom data](https://surprise.readthedocs.io/en/stable/getting_started.html#use-a-custom-dataset)

```python
from surprise import accuracy, Dataset, Reader, SVD
import os

reader = Reader()
data = Dataset.load_from_df(user_ratings[['user_id', 'movie_id', 'rating']], reader=reader)
```

Sử dụng K-Fold để huấn luyện và đánh giá mô hình


Tham khảo đường dẫn sau: [Use cross-validation iterators
](https://surprise.readthedocs.io/en/stable/getting_started.html#use-cross-validation-iterators)


```python
from surprise.model_selection import KFold, cross_validate

kfold = KFold(random_state=42, n_splits=7)
svd = SVD()
for X_train, X_test in kfold.split(data):
    svd.fit(X_train)
    y_pred = svd.test(X_test)
    accuracy.rmse(y_pred, verbose=3)
    accuracy.mse(y_pred, verbose=3)
```
